**Análise de Loja de Vendas**

**Importando bibliotecas**

```python
import pandas as pd

df = pd.DataFrame({
    'produto': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'categoria': ['Eletrônico', 'Eletrônico', 'Eletrônico',
                  'Roupas', 'Roupas', 'Roupas',
                  'Comida', 'Comida', 'Comida'],
    'vendas': [1200, 800, 500, 900, 200, 300, 3000, 1500, 1000],
    'lucro': [300, 150, 90, 250, 40, 70, 800, 350, 200]
})
```

**Visualização de dados iniciais**

```python
df.head()
```

```python
df.tail()
```

```python
## df.rename(columns={'produto

df.columns
df = df.rename(columns={'produto': 'Produto', 
                        'categoria': 'Categoria',
                        'vendas': 'Vendas',
                        'lucro': 'Lucro'})
```

```python
# Categorias que ocorreram vendas acima de R$ 1.000,00
df.loc[df['Vendas'] > 1000]
```

```python
# Verificar apenas produtos + vendas + lucro
df.loc[:, ['Produto', 'Vendas', 'Lucro']]
```

```python
# Verificar o somatório de vendas por categoria
df.groupby('Categoria')['Vendas'].sum()
```

```python
# Categorias com maior quantidade de vendas
df.sort_values(by='Vendas', ascending=False).head(3)
```

```python
# Soma de lucro por produto
df.groupby('Produto')['Lucro'].sum()
```

```python
# Produtos que menos venderam
df.sort_values(by='Vendas', ascending=True).head(3)
```

```python
# Produtos com menos vendas
df.sort_values(by='Vendas').head(3)

edge = df['Vendas'].quantile(0.50)
df.loc[df['Vendas'] > edge]
```

**Insights**

Categoria que mais vende? **Comida**

Produto que mais vende? **Produto A**

Onde estão os "outliers" das vendas? **Concentrados na categoria Comida, pois ultrapassa R$ 3.000,00**

Quais produtos precisam de atenção? **Produtos B e C; setores Roupas e Eletrônico respectivamente**
